Importando as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Bibliotecas importadas com sucesso!")
```

Importando os dados

```python
dados = pd.read_csv("dados.csv", sep=';')
print(f"As dimensões da base de dados é {dados.shape}")
dados
```

Visualização das colunas do conjunto de dados

```python
# PIB
sns.histplot(dados.PIB, kde=True, bins=10).set(title='Valores de PIB')
```

```python
# Valores de Empenho
sns.histplot(dados.VALOREMPENHO, kde=True, bins=10).set(title='Valores de Empenho')
```

```python
# PIB por municípios
tmp = dados.sort_values('PIB').head(10)
tmp = tmp.iloc[:,1:3]  # Selecionando as colunas de 1 a 3
tmp
tmp.plot.bar(x='MUNICIPIO', y='PIB', color='gray')
```

```python
# Empenho por municípios
tmp = dados.sort_values('VALOREMPENHO').head(10)
tmp = tmp.iloc[:,[1,3]]
tmp
tmp.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'gray')
```
